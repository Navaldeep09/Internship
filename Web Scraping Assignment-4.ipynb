# Question 1

```python
#opening chrome
driver=webdriver.Chrome()
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
# fetching details
Name=[]
Artist=[]
Date=[]
Views=[]

#Fetching details 
try:
    name_tag = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[1]')
    for n in name_tag:
        Name.append(n.text)
except NoSuchElementException:
        Name.append('-')
        
try:
    artist_tag = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
    for a in artist_tag:
        Artist.append(a.text)
except NoSuchElementException:
        Artist.append('-')
        
        
try:
    date_tag = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
    for d in date_tag:
        Date.append(d.text)
except NoSuchElementException:
        Date.append('-')
        

try:
    views_tag = driver.find_elements(By.XPATH, '//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
    for v in views_tag:
        Views.append(v.text)
except NoSuchElementException:
        Views.append('-')
        
        

#creating Dataframe

df=pd.DataFrame({'Name':Name,'Artist':Artist,'Date':Date,'Views':Views})
df
        
```

```python
#opening chrome
driver=webdriver.Chrome()
driver.get("https://www.imdb.com/list/ls512407256/")
```

```python
Name=[]
Year=[]
Genre=[]
Time=[]
Ratings=[]
Votes=[]

#Fetching details 
try:
    name_tag = driver.find_elements(By.XPATH, '//h3[@class="lister-item-header"]/a')
    for n in name_tag:
        Name.append(n.text)
except NoSuchElementException:
    Name.append('-')
    

try:
    year_tag = driver.find_elements(By.XPATH, '//h3[@class="lister-item-header"]/span[2]')
    for y in year_tag:
        Year.append(y.text)
except NoSuchElementException:
    Year.append('-')
    
    
try:
    genre_tag = driver.find_elements(By.XPATH, '//p[@class="text-muted text-small"]/span[5]')
    for g in genre_tag:
        Genre.append(g.text)
except NoSuchElementException:
    Genre.append('-')
    
    
try:
    time_tag = driver.find_elements(By.XPATH, '//p[@class="text-muted text-small"]/span[3]')
    for t in time_tag:
        Time.append(t.text)
except NoSuchElementException:
    Time.append('-')
    
try:
    rating_tag = driver.find_elements(By.XPATH, '//div[@class="ipl-rating-star small"]/span[2]')
    for r in rating_tag:
        Ratings.append(r.text)
except NoSuchElementException:
    Ratings.append('-')
    
try:
    votes_tag = driver.find_elements(By.XPATH, '//div[@class="lister-item-content"]/p[4]/span[2]')
    for v in votes_tag:
        Votes.append(v.text)
except NoSuchElementException:
    Votes.append('-')

    
#creating Dataframe

df=pd.DataFrame({'Name':Name,'Year':Year,'Genre':Genre,'Time':Time,'Ratings':Ratings,'Votes':Votes})
df
    
```

# Question 8

```python
#opening chrome
driver=webdriver.Chrome()
driver.get("https://archive.ics.uci.edu/")
dataset=driver.find_element(By.XPATH,'//ul[@class="menu menu-horizontal"]/li[1]/a')
dataset.click()
```

```python
Dataset_Name=[]
Data_Type=[]
Task=[]
Instances=[]
Attributes=[]
Year=[]

#Fetching details 
try:
    name_tag = driver.find_elements(By.XPATH, '//h2[@class="truncate text-primary"]/a')
    for n in name_tag:
        Dataset_Name.append(n.text)
except NoSuchElementException:
    Dataset_Name.append('-')
    
    
try:
    type_tag = driver.find_elements(By.XPATH, '//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[2]/span')
    for t in type_tag:
        Data_Type.append(t.text)
except NoSuchElementException:
    Data_Type.append('-')  
    
try:
    task_tag = driver.find_elements(By.XPATH, '//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[1] /span')
    for tk in task_tag:
        Task.append(tk.text)
except NoSuchElementException:
    Task.append('-')  

try:
    instances_tag = driver.find_elements(By.XPATH, '//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[3]/span')
    for i in instances_tag:
        Instances.append(i.text)
except NoSuchElementException:
    Instances.append('-')  

try:
    attribute_tag = driver.find_elements(By.XPATH, '//div[@class="my-2 hidden gap-4 md:grid grid-cols-12"]/div[4]/span')
    for a in attribute_tag:
        Attributes.append(a.text)
except NoSuchElementException:
    Attributes.append('-')  
                

#creating Dataframe

df=pd.DataFrame({'Dataset name':Dataset_Name,'Data type':Data_Type,'Task':Task,'No of instances':Instances,'No of attribute':Attributes})
df
  
```
